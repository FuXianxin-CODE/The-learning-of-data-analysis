# 一、读取CSV文件代码

```python
import csv

# 获取数据的时候通过下标来获取
with open('stock.csv','r') as fp:
    reader = csv.reader(fp)
    titles = next(reader)
    for x in reader:
        print(x)
```

```python
# 如果想要在获取数据的时候通过标题来获取，可以使用DictReader
with open("stock.csv",'r') as fp:
    reader = csv.DictReader(fp)
    for x in reader:
        print(x['turnoverVol'])
```

# 写入数据到csv文件

```python
# 创建writer对象，利用writerow写入一行，利用writerows写入多行
import csv

headers = ['name','age','classroom']
values = [
    ('九天雪',23,'021班'),
    ('小白',20,'019班'),
    ('雷霆',13,'025班')
]
with open('test.csv','w',newline='') as fp:
    writer = csv.writer(fp)
    writer.writerow(headers)
    writer.writerows(values)
```

```python
# 也可以使用字典的方式把数据写入，需要使用DictWriter
headers = ['name','age','classroom']
values = [
    {"name":'肥肥',"age":20,"classroom":'222'},
    {"name":'胖胖',"age":30,"classroom":'333'}
]
with open('test.csv','w',newline='') as fp:
    writer = csv.DictWriter(fp,headers)
    writer.writerow({'name':"小胖子","age":24,"classroom":'001'})
    writer.writerows(values)
```
